Symbol: The stock symbol/name or the identifier of the security.

Series: The series of the security, such as equity shares, debentures, etc. Values are EQ, BE, BL, BT, GC, IL and IQ.

Open: Price at which the security opened on the day of the report.

High: Highest traded price of the security during the day.

Low: Lowest traded price of the security during the day.

Close: Price at which the security was traded at the end of the market.

Last: Price at which the security was traded at the last transaction before the close of the market.

Prev Close: Closing price of the security on the previous trading day.

TOTTRDQTY: The total quantity or number of shares traded during the day.

TOTTRDVAL: The total value of the shares traded, calculated as the product of the traded quantity and the last traded price.

TIMESTAMP: The date of the report.

TOTALTRADES: The number of trades that were executed for the security during the day. E.g. for a stock, Trade 1 has quantity = 100, Trade 2 quantity = 1000. TOTALTRADES = 2 and TOTTRDQTY = 1100

ISIN: International Securities Identification Number. Identify the security across exchanges using this ID.

```python
import pandas as pd
from datetime import date, timedelta
import nselib
from nselib import capital_market
from collections import namedtuple
from calendar import monthrange
from time import time

year = 2023
months = [x for x in range(1, 13)]
# sdate = '01-01-2023'
# edate = '31-12-2023'
for month in months:

    num_days = monthrange(year, month)[1]
    print(num_days)
    sdate = date(2023, month, 1)  # start date
    edate = date(2023, month, num_days)  # end date
    dates = pd.date_range(sdate, edate, freq="d").strftime("%d-%m-%Y").tolist()
    print(dates)
    bhav_df = pd.DataFrame()

    for date in dates:
        try:
            t_start = time()
            # print(f'Calling bhav_copy_with_delivery with trade_date = {date}')
            temp_df = capital_market.bhav_copy_with_delivery(trade_date=date)
            bhav_df = pd.concat([bhav_df, temp_df], ignore_index=True)
            t_end = time()
            print(f"Took %.3f second for {date}" % (t_end - t_start))
        except KeyError:
            pass
    print(f"Print complete for {month}")
    print(bhav_df.shape)
print(f"Print complete for {year}")
```

```python
import pandas as pd
from datetime import date, timedelta
import nselib
from nselib import capital_market
from collections import namedtuple
from calendar import monthrange
from time import time


def get_dates(year, month=None):
    dates = {}
    months = [x for x in range(1, 13)]
    for month in months:
        # month_start = time()
        num_days = monthrange(year, month)[1]
        sdate = date(year=year, month=month, day=1)  # start date
        edate = date(year=year, month=month, day=num_days)  # end date
        dates[month] = (
            pd.date_range(sdate, edate, freq="d").strftime("%d-%m-%Y").tolist()
        )
        # bhav_df = pd.DataFrame()
    return dates
```
